```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)  
PYTHONHASHSEED = 0
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix, recall_score, precision_score
import tensorflow as tf

%matplotlib inline
```

```python
# training
train_df = pd.read_csv("./data/PM_train.txt", sep=" ", header=None)

train_df.drop(train_df.columns[[26, 27]], axis=1, inplace=True)
train_df.columns = ['id', 'cycle', 'setting1', 'setting2', 'setting3', 's1', 's2', 's3',
                     's4', 's5', 's6', 's7', 's8', 's9', 's10', 's11', 's12', 's13', 's14',
                     's15', 's16', 's17', 's18', 's19', 's20', 's21']
```

```python
# test
test_df = pd.read_csv('./data/PM_test.txt', sep=" ", header=None)
test_df.drop(test_df.columns[[26, 27]], axis=1, inplace=True)
test_df.columns = ['id', 'cycle', 'setting1', 'setting2', 'setting3', 's1', 's2', 's3',
                     's4', 's5', 's6', 's7', 's8', 's9', 's10', 's11', 's12', 's13', 's14',
                     's15', 's16', 's17', 's18', 's19', 's20', 's21']
```

```python
# ground truth
truth_df = pd.read_csv('./data/PM_truth.txt', sep=" ", header=None)
truth_df.drop(truth_df.columns[[1]], axis=1, inplace=True)
```

```python
train_df = train_df.sort_values(['id','cycle'])
train_df.head()
```

# Data Labeling

```python
# create and add column RUL
med = pd.DataFrame(train_df.groupby('id')['cycle'].max()).reset_index()
med.columns = ['id', 'max']
train_df = train_df.merge(med, on=['id'], how='left')
train_df['RUL'] = train_df['max'] - train_df['cycle']
train_df.drop('max', axis=1, inplace=True)
train_df.head()
```

```python
# add labels for binary/multiclass classifications
w1 = 30
w0 = 15
train_df['label1'] = np.where(train_df['RUL'] <= w1, 1, 0 )
train_df['label2'] = train_df['label1']
train_df.loc[train_df['RUL'] <= w0, 'label2'] = 2
train_df.head()
```

# Data Normalization

```python
train_df['cycle_norm'] = train_df['cycle']
cols_to_normalize = train_df.columns.difference(['id','cycle','RUL','label1','label2'])
min_max = preprocessing.MinMaxScaler()
norm_train_df = pd.DataFrame(min_max.fit_transform(train_df[cols_to_normalize]), 
                             columns=cols_to_normalize, 
                             index=train_df.index)
join_df = train_df[train_df.columns.difference(cols_to_normalize)].join(norm_train_df)
train_df = join_df.reindex(columns = train_df.columns)
train_df.head()
```

# Prepare test data

```python
test_df['cycle_norm'] = test_df['cycle']
norm_test_df = pd.DataFrame(min_max.transform(test_df[cols_to_normalize]), 
                            columns=cols_to_normalize, 
                            index=test_df.index)
test_join_df = test_df[test_df.columns.difference(cols_to_normalize)].join(norm_test_df)
test_df = test_join_df.reindex(columns = test_df.columns)
test_df = test_df.reset_index(drop=True)
test_df.head()
```

```python
# use ground truth dataset to generate column max for test data

helpcols = pd.DataFrame(test_df.groupby('id')['cycle'].max()).reset_index()
helpcols.head(100)
```

```python
helpcols.columns = ['id', 'max']
truth_df.columns = ['more']
truth_df['id'] = truth_df.index + 1
truth_df['max'] = helpcols['max'] + truth_df['more']
truth_df.drop('more', axis=1, inplace=True)

truth_df.head(5)
```

```python
# create RUL for test data

test_df = test_df.merge(truth_df, on=['id'], how='left')
test_df['RUL'] = test_df['max'] - test_df['cycle']
test_df.drop('max', axis=1, inplace=True)
test_df.head()
```

```python
# create columns with labels 'label1' for binary classification(if defect in next w1 cycles) and for multiclass classification 
# 'label2' (if defect in next w0 cycles) for test data

test_df['label1'] = np.where(test_df['RUL'] <= w1, 1, 0 )
test_df['label2'] = test_df['label1']
test_df.loc[test_df['RUL'] <= w0, 'label2'] = 2
test_df.head()
```

# Prepairing data for modelling

```python
# pick a large window size of 50 cycles
sequence_length = 50
```

```python
# preparing data for visualizations 
# window of 50 cycles prior to a failure point for engine with a passed id number
def prepare_data_visual(test_df, id_num):
    engine_id = test_df[test_df['id'] == id_num]
    engine_id_50cycleWindow = engine_id[engine_id['RUL'] <= engine_id['RUL'].min() + 50]
    cols1 = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9', 's10']
    engine_id_50cycle1 = engine_id_50cycleWindow[cols1]
    cols2 = ['s11', 's12', 's13', 's14', 's15', 's16', 's17', 's18', 's19', 's20', 's21']
    engine_id_50cycle2 = engine_id_50cycleWindow[cols2]
    return (engine_id_50cycle1, engine_id_50cycle2)
```

```python
engine_id4_50cycle1, engine_id4_50cycle2 = prepare_data_visual(test_df, 4)
```

```python
# plotting sensor data for engine ID 4 prior to a failure point - sensors 1-10 
ax1 = engine_id4_50cycle1.plot(subplots=True, sharex=True, figsize=(20,20))
```

```python
# plotting sensor data for engine ID 4 prior to a failure point - sensors 11-21 
ax2 = engine_id4_50cycle2.plot(subplots=True, sharex=True, figsize=(20,20))
```

```python
# Keras LSTM layers expect an input in the shape of a numpy array of 3 dimensions (samples, time steps, features) 
# where:
# 1. Dimension - samples is the number of training sequences 
# 2. Dimension - time steps is the look back window or sequence length 
# 3. Dimension - features is the number of features of each sequence at each time step.

def gen_sequence(id_df, seq_length, seq_cols):
    """ Only sequences that meet the window-length are considered, no padding is used. This means for testing
    we need to drop those which are below the window-length. An alternative would be to pad sequences so that
    we can use shorter ones """
    data_array = id_df[seq_cols].values
    num_elements = data_array.shape[0]
    for start, stop in zip(range(0, num_elements-seq_length), range(seq_length, num_elements)):
        yield data_array[start:stop, :]
```

```python
# take the feature columns 
sensor_cols = ['s' + str(i) for i in range(1,22)]
sequence_cols = ['setting1', 'setting2', 'setting3', 'cycle_norm']
sequence_cols.extend(sensor_cols)
```

```python
# generator for the sequences
seq_generated = (list(gen_sequence(train_df[train_df['id']==id], sequence_length, sequence_cols)) 
           for id in train_df['id'].unique())
```

```python
# generate sequences and convert to numpy array
seq_array = np.concatenate(list(seq_generated)).astype(np.float32)
seq_array.shape
```

```python
# generate labels
def gen_labels(id_df, seq_length, label):
    data_array = id_df[label].values
    num_elements = data_array.shape[0]
    return data_array[seq_length:num_elements, :]
```

```python
label_gen = [gen_labels(train_df[train_df['id']==id], sequence_length, ['label1']) 
             for id in train_df['id'].unique()]
label_array = np.concatenate(label_gen).astype(np.float32)
label_array.shape
```

# Prepairing test data

```python
# only the last cycle data for each engine id in the test data is kept for testing purposes. 
# In order to compare the results to the template, we pick the last sequence for each id in the test data.

seq_test_last = [test_df[test_df['id']==id][sequence_cols].values[-sequence_length:] 
                       for id in test_df['id'].unique() if len(test_df[test_df['id']==id]) >= sequence_length]

seq_test_last = np.asarray(seq_test_last).astype(np.float32)
seq_test_last.shape
```

```python
len_mask = [len(test_df[test_df['id']==id]) >= sequence_length for id in test_df['id'].unique()]
```

```python
label_test_last = test_df.groupby('id')['label1'].nth(-1)[len_mask].values
label_test_last = label_test_last.reshape(label_test_last.shape[0],1).astype(np.float32)
label_test_last.shape
```

```python
print(seq_test_last.shape)
print(label_test_last.shape)
```

# LSTM Network

# the network has a following layers:
# 1. an LSTM layer with 100 units
# 2. dropout after each LSTM layer to control overfitting
# 3. an LSTM layer with 50 units
# 4. dropout after each LSTM layer to control overfitting
# 5. a Dense output layer with single unit and sigmoid activation since this is a binary classification

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense
```

```python
# build the network
nb_features = seq_array.shape[2]
nb_out = label_array.shape[1]

model = Sequential()

model.add(LSTM(
         input_shape=(sequence_length, nb_features),
         units=100,
         return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(
          units=50,
          return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(units=nb_out, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
from datetime import datetime

from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model
```

```python
# Paths

PATH_DATA = "./"
PATH_MODELS = PATH_DATA + "models/"
PATH_CHECKPOINTS = PATH_MODELS + "checkpoints/"
```

```python
# stop training when validation acc starts dropping
# save checkpoint of model each period
now = datetime.now().strftime("%Y-%m-%d_%H%M")
# create callbacks
callbacks = [
    EarlyStopping(monitor="val_loss", verbose=1, patience=2),
    ModelCheckpoint(
        PATH_CHECKPOINTS + now + "_Model_{epoch:02d}_{val_loss:.4f}.h5",
        monitor="val_loss",
        save_best_only=True,
        verbose=1,
    ),
]
```

```python
%%time
# train the network
hist = model.fit(seq_array, label_array, epochs=10, batch_size=200, validation_split=0.05, verbose=1, callbacks = callbacks)
```

```python
import seaborn as sns
```

```python
loss = pd.DataFrame(
    {"train loss": hist.history["loss"], "test loss": hist.history["val_loss"]}
).melt()
loss["epoch"] = loss.groupby("variable").cumcount() + 1
sns.lineplot(x="epoch", y="value", hue="variable", data=loss).set(
    title="Model loss", ylabel=""
)
```

```python
# training metrics
scores = model.evaluate(seq_array, label_array, verbose=1, batch_size=200)
print('Accurracy: {}'.format(scores[1]))
```

```python
# make predictions and compute confusion matrix
y_pred = model.predict(seq_array,verbose=1, batch_size=200)
pred_class = (y_pred > 0.5).astype(int)
y_true = label_array
print(y_true)
pred_class
```

```python
print('Confusion matrix\n- x-axis is true labels.\n- y-axis is predicted labels')
cm = confusion_matrix(y_true, pred_class)
cm
```

```python
# compute precision and recall
precision = precision_score(y_true, pred_class)
recall = recall_score(y_true, pred_class)
print( 'precision = ', precision, '\n', 'recall = ', recall)
```
